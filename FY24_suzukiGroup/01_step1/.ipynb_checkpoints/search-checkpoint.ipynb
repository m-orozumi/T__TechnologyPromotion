```python
## ライブラリのインポート ##
# スクレイピング用
from time import sleep

import requests
from bs4 import BeautifulSoup

# データ保存用
import pandas as pd



## スクレイピング ##
# 対象サイトのURLを格納する
#url = "https://zdh.stagingbridge.net/"
url = "https://zdh.stagingbridge.net/?s=AWS"

# requestsを使用してhtmlを取得　→変数rに格納
#r = requests.get(url, params={'q': 'AWS'})
r = requests.get(url)

# 対象サイトへ負荷をかけないように、1sec待機
sleep(1)

# 取得結果を解析してsoupに格納
soup = BeautifulSoup(r.content, "html.parser")

# 本文が入っている「txt-link txt-nomal」クラスを抽出
　#URLにアクセスしhtmlを取得　→取得対象のタイトル・URLが記載されているタグとクラスを確認して指定
contents = soup.find_all("a", class_="txt-link txt-nomal")
#contents　#動作確認用

# 本文を抽出
#data_list = contents.find_all("a")
#contents[0].contents[0]
#contents[0].attrs['href']
#for content in contents:

# データフレームに格納するため、抽出結果をまとめる
data_result = []

for data in contents:
    # 抽出した本文をデータセットに格納する
    data_set = {
         'タイトル': data.contents
        ,'URL': data.attrs['href']
    }

    # 取得したデータセットをd_listに格納する
    data_result.append(data_set)



## データを保存 ##
# data_resultを、データフレームに格納
df = pd.DataFrame(data_result)

# to_csv()を使って、データフレームをCSV出力する
df.to_csv("search.csv",index=None,encoding="utf-8-sig")
```
